```python
df1["Visiting Behaviour"] = (df1['D6']=="Yes")*1
```

```python
df2=df1.drop('D6',axis=1)
```

```python
# Creating target variable - for data modelling
Y=df2[['Visiting Behaviour']]
```

```python
# Creating a list of all independent variable except Visiting Behaviour
```

```python
Independent_Variables =['OIQ1', 'OIQ2', 'OIQ3', 'UFA1', 'UFA2',
       'UFA3', 'S1', 'S2', 'S3', 'S4', 'IVTD1', 'IVTD2', 'IVTD3', 'DMI1',
       'DMI2', 'EWOM1', 'EWOM2']
```

```python
X=df2[['OIQ1', 'OIQ2', 'OIQ3', 'UFA1', 'UFA2',
       'UFA3', 'S1', 'S2', 'S3', 'S4', 'IVTD1', 'IVTD2', 'IVTD3', 'DMI1',
       'DMI2', 'EWOM1', 'EWOM2']]
```

```python
X.columns
```

```python
X = X.apply(LE.fit_transform)
```

```python
X.columns
```

```python
# MODEL TRAINING
```

```python
# Splitting data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=8)
```

```python
# Fitting Decision Tree model on data
VB = DecisionTreeClassifier(max_leaf_nodes=10)
```

```python
VB.fit(X_train,Y_train)
```

```python
VB.score(X_train,Y_train)
```

```python
# Predicting values on test set
Y_Pred = VB.predict(X_test)
```

```python
Y_Pred
```

```python
# Checking test accuracy
Accuracy = accuracy_score(Y_test,Y_Pred)*100
```

```python
Accuracy
```

```python
# Import the necessary module
from sklearn.metrics import confusion_matrix

```

```python
confusion_matrix(Y_test,Y_Pred)
```
